```python
import torch
from torch import nn
import matplotlib.pyplot as plt


#check pytorch version
print(torch.__version__)
```

## Data (Preparing and loading)

```python
# Create known parameters
weight = 0.7
bias = 0.3

start = 0
end = 1
step = 0.02

X = torch.arange(start, end, step).unsqueeze(dim=1)
y = weight * X + bias

X[:10], y[:10], len(X), len(y)
```

```python
### Split data into training and test sets

train_split = int(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split] # takes everything up to train_split
X_test, y_test = X[train_split:], y[train_split:] # takes everything from train_split onwards

len(X_train), len(y_train), len(X_test), len(y_test)
```

```python
def plot_predictions(X_train, y_train, X_test, y_test, y_pred):
    plt.figure(figsize=(10, 7))
    plt.scatter(X_train, y_train, s=15, color='blue', label='Training data')
    plt.scatter(X_test, y_test, s=15, color='green', label='Testing data')
    plt.scatter(X_test, y_pred, s=15, color='red', label='Predictions')
    plt.legend()
    plt.show()

plot_predictions(X_train, y_train, X_test, y_test, y_test)
```

```python
## My first pytorch model

class LinearRegression(nn.Module):
    def __init__(self):
        super().__init__()

        #initialize the weights and biases with random values
        self.weights = nn.Parameter(torch.randn(1, requires_grad=True, dtype=torch.float))
        self.bias = nn.Parameter(torch.randn(1, requires_grad=True, dtype=torch.float))

        # Forward method to define the computation in the model
        def forward(self, x: torch.Tensor) -> torch.Tensor: # x is the input tensor
            return self.weights * x + self.bias #linear regression formula
```
